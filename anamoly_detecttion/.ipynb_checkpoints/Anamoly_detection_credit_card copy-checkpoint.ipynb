# Introduction 

* In the real world, fraud often goes undiscovered, and only the fraud that is caught provides any labels for the datasets. 

* Moreover, fraud patterns change over time, so supervised systems that are built using fraud labels become stale, capturing historical patterns of fraud but failing to adapt to newly emerging patterns.

* For these reasons (the lack of sufficient labels and the need to adapt to newly emerging patterns of fraud as quickly as possible), unsupervised learning fraud detection systems are in vogue.

* In this notebook, we will build such a solution using PCA 

## Step 1: Standardize the data
* For example, if X has centered values [x1-μ, x2-μ, ..., xn-μ] and standard deviation σ, then the standardized values of X are [(x1-μ)/σ, (x2-μ)/σ, ..., (xn-μ)/σ].
* Repeat this process for each variable in the dataset.

## Step 2: Compute the covariance matrix
* The covariance between two variables X and Y is given by the formula:
* cov(X,Y) = sum((Xi - mean(X))(Yi - mean(Y))) / (n-1), where Xi and Yi are the ith observations of X and Y, respectively, and mean(X) and mean(Y) are the means of X and Y, respectively.

## Step 3: Compute the eigenvectors and eigenvalues of the covariance matrix
* Solve the following equation for the eigenvectors v and corresponding eigenvalues λ:
* (C - λI) v = 0, where C is the covariance matrix, I is the identity matrix, and v is a non-zero vector.
* This equation is known as the eigenvector equation and can be rearranged to (C - λI)^(-1)v = 0.
* The solutions to this equation are the eigenvectors and eigenvalues of the covariance matrix.
* The magnitude of an eigenvalue λ represents the amount of variance explained by the corresponding eigenvector v.
* The total variance in the data is given by the sum of the eigenvalues: var(X) = sum(λ).
* The variance explained by the kth principal component is given by (λk / var(X)).

## Step 4: Choose the principal components
* Rank the eigenvalues in decreasing order.
* Select the top k eigenvalues, where k is the desired number of principal components.
* The corresponding eigenvectors are the loading vectors for the k principal components.
* Each loading vector represents the weightings of each variable on the corresponding principal component.
* The loading vectors can be normalized to have unit length by dividing each eigenvector by its corresponding eigenvalue.

## Step 5: Transform the data

* Multiply the standardized data matrix by the matrix of selected eigenvectors to get the new coordinates of the data in the reduced-dimensional space.
* This new coordinate system is orthogonal, meaning that the principal components are uncorrelated with each other.
* The coordinates of each observation in the new coordinate system are called the principal component scores.
* The variance of each principal component score is equal to the corresponding eigenvalue.


Let's consider a small data table with two variables X and Y, and three data points:

|Data Point|X|Y|
| --- | --- | --- |
|1|2|3|
|2|3|5|
|3|4|7|

To calculate the principal components of this dataset, we first need to calculate the covariance matrix. The covariance between two variables X and Y can be calculated using the following formula:

`cov(X,Y) = sum((Xi - mean(X))(Yi - mean(Y))) / (n-1) `

`cov(X,Y) = ((2-3)(3-5) + (3-3)(5-5) + (4-3)(7-5)) / 2 = -1`

Similarly, we can calculate the variance of X and Y using the following formula:

`var(X) = sum((Xi - mean(X))^2) / (n-1)`

`var(Y) = sum((Yi - mean(Y))^2) / (n-1)`

Using this formula, we can calculate the variance of X and Y for our dataset:

`var(X) = ((2-3)^2 + (3-3)^2 + (4-3)^2) / 2 = 1.33`

`var(Y) = ((3-5)^2 + (5-5)^2 + (7-5)^2) / 2 = 3.08`

## Calculate Eigen Values
Now that we have the covariance matrix and the variances, we can calculate the eigenvalues and eigenvectors. The eigenvalues represent the amount of variance in the data explained by each principal component, while the eigenvectors represent the direction of each principal component.

To calculate the principal components, we first need to calculate the covariance matrix:

|    | X    | Y    |
|----|------|------|
| X  | 1.33 | -1   |
| Y  | -1   | 3.08 |


To calculate the eigenvectors, we need to find the values of x and y that satisfy the equation:

`(C - λI)v = 0`

Where v is the eigenvector.
and I is identity matrix
The characteristic equation becomes:

`det(C - λI) = (1.33 - λ)(3.08 - λ) - (-1)(-1) = λ^2 - 4.41λ + 4.84 = 0`

Solving for λ, we get:

`= (1.33 - λ)(3.08 - λ) - (-1)(-1)    // expand the determinant along the first row`

`= λ^2 - (1.33 + 3.08)λ + (1.33*3.08 - (-1)*(-1))`

`= λ^2 - 4.41λ + 4.84`

`λ = (-b ± sqrt(b^2 - 4ac)) / (2a)`

`where a = 1, b = -4.41, and c = 4.84`

`λ1 = (4.41 - sqrt(4.41^2 - 4*1*4.84)) / (2*1) = (4.41 - sqrt(8.3921)) / 2`
   `= 0.58`

`λ2 = (4.41 + sqrt(4.41^2 - 4*1*4.84)) / (2*1) = (4.41 + sqrt(8.3921)) / 2`

   `= 3.83`

`λ1 = 0.58`

`λ2 = 3.83`
These are the eigenvalues for our dataset.

The eigenvalues represent the amount of variance in the data explained by each principal component. We can calculate the proportion of the total variance explained by each principal component by dividing the eigenvalue by the sum of all eigenvalues:

`Proportion of variance explained by PCA1 = λ1 / (λ1 + λ2) = 0.13`

`Proportion of variance explained by PCA2 = λ2 / (λ1 + λ2) = 0.87`

So PCA2 explains more variance in the data than PCA1.

`-2.5x - y = 0`

Solving for y, we get:

`y = -2.5x`
`x = -.75y`

EigenVector=v2= `[-2.5, 1]`
and the elements inside the eigen vector are called component loadings

## Methodology

* We take the difference between the original vs the pca dataframe which we created from pcs components using inverse transformation
* then we transform the difference using min max scaller from 0 to 1 scale
* and any values above .5 are anamolous and below 0.1 are non anomolous
* we can consider this score as probability score and use this against the labels to calculate, precision, recall and threshhold using sklearn precision recall curve metric. 
* we can all also calculate average precision score using the average precision score metric.


# Define evaluation functions

```python
# Calculate reconstruction error
def anomalyScores(originalDF, pca_df):
    loss = ((originalDF.values- pca_df.values)**2).sum(axis=1)    
    loss = pd.Series(data=loss,index=originalDF.index)
    loss = (loss-np.min(loss))/(np.max(loss)-np.min(loss))
    return loss
```

## PCA custom class

```python
from sklearn.base import BaseEstimator, TransformerMixin
class pca_anom(BaseEstimator, TransformerMixin):
    def __init__(self, variance= .97):   
        self.variance = variance
        self.model = None
    
    def fit(self, X,y=None):       
       self.model= PCA(n_components= self.variance, whiten=False, random_state=2018)
       self.model= self.model.fit(X)
       return self
    
    def transform(self, X, y=None):                
        self.columns = self.model.transform(X).shape[1]        
        return self.model.transform(X)
    
    def get_feature_names_out(self, names=None):        
        return [f'pca{x}'for x in range(self.columns) ]    
    
    def predict(self, X ):                   
        df_pca_inv= pd.DataFrame(self.model.inverse_transform(self.model.transform(X)))
        df_pca_inv.columns= X.columns       
        results= pd.DataFrame()        
        results["prob"]= anomalyScores(X,df_pca_inv)
        anamloies= results.sort_values("prob" , ascending= False).head(y_train.value_counts()[1]+20)        
        results= results.assign(class_=  np.where(results.index.isin(anamloies.index), 1,0 ))
        return results[["class_"]]
    
    def predict_proba(self, X ):           
        
        df_pca_inv= pd.DataFrame(self.model.inverse_transform(self.model.transform(X)))
        df_pca_inv.columns= X.columns       
        results= pd.DataFrame()        
        results["prob"]= anomalyScores(X,df_pca_inv)
        return results
```

```python
from sklearn.base import MetaEstimatorMixin, clone
from sklearn.base import BaseEstimator, TransformerMixin

class pca_anom(BaseEstimator, TransformerMixin):
    def __init__(self,estimator):           
        self.estimator = estimator
    
    def fit(self, X,y=None):       
       estimator_= clone(self.estimator)
       self.model= estimator_.fit(X)
       return self   
        
    def transform(self, X, y=None):                
        self.columns = self.model.transform(X).shape[1]        
        return self.model.transform(X)
    
    def get_feature_names_out(self, names=None):        
        return [f'pca{x}'for x in range(self.columns) ]    
    
    def predict(self, X ):                   
        df_pca_inv= pd.DataFrame(self.model.inverse_transform(self.model.transform(X)))
        df_pca_inv.columns= X.columns       
        results= pd.DataFrame()        
        results["prob"]= anomalyScores(X,df_pca_inv)
        anamloies= results.sort_values("prob" , ascending= False).head(y_train.value_counts()[1]+20)        
        results= results.assign(class_=  np.where(results.index.isin(anamloies.index), 1,0 ))
        return results[["class_"]]
    
    def predict_proba(self, X ):             
        df_pca_inv= pd.DataFrame(self.model.inverse_transform(self.model.transform(X)))
        df_pca_inv.columns= X.columns       
        results= pd.DataFrame()        
        results["prob"]= anomalyScores(X,df_pca_inv)
        return results
```

```python
from sklearn.decomposition import FastICA
pipe= Pipeline(steps=[
    ("preprocess", prep),    
    ('model', pca_anom(PCA(n_components= 30, random_state=2018)))        
    ]).set_output(transform="pandas")

pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, recall_score ,average_precision_score, precision_recall_curve, precision_score, roc_auc_score,roc_curve, auc 
def eval_scores(actuals, pred):
    res= pd.DataFrame({ 
    "accuracy": [accuracy_score(actuals, pred)], 
    "recall": [recall_score(actuals, pred)],
    "roc_auc": [roc_auc_score(actuals, pred)],
    "precision": [precision_score(actuals, pred)]})    
    return res
models= {}
scores= {}
for j in range(20,30):
    models["pca"+str(j)]= pipe.set_params(**{"model__estimator__n_components": j})
    models["pca"+str(j)].fit(X_train)
    scores["pca"+str(j)]= eval_scores(y_train, models["pca"+str(j)].predict(X_train) )
```
